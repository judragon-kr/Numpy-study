```python
import numpy as np
```

```python
# Scalar arithmetic
array = np.array([1, 2, 3])
print(array + 1)
print(array - 2)
print(array * 2)
print(array / 2)
print(array ** 2)
```

```python
# Vertorized math funcs
array = np.array([1.01, 2.5, 3.99])
print(np.sqrt(array))
print(np.round(array))
print(np.pi)
```

```python
# EXERCISE
radii = np.array([1, 2, 3])
print(np.pi * radii ** 2)
```

```python
# Elemet-wise arithmetic
array0 = np.array([1, 2, 3])
array1 = np.array([4, 5, 6])
print(array0 + array1)
print(array0 - array1)
print(array0 * array1)
print(array0 / array1)
print(array0 ** array1)
```

```python
# Comparison operators
scores = np.array([91, 55, 100, 73, 83, 74])
print(scores == 100)
print(scores >= 60)
print(scores < 60)

scores[scores < 60] = 0
print(scores)
```
