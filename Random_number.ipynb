```python
import numpy as np
```

```python
# 마크랑 똑같이 시드가 같으면 같은 수가 나옴 
rng = np.random.default_rng()

# 정수
print(rng.integers(1, 7, size=(2, 10)))

# uniform, uniform distribution
print(rng.uniform(-1, 7, size=(2, 10)))

# Shuffle
array = np.array([1, 2, 3, 4, 5])
rng.shuffle(array)
print(array)

# Choice
fruits = np.array(["apple", "orange", "banana", "coconut", "pineapple"])
fruit = rng.choice(fruits, size=3)
print(fruit)
fruit = rng.choice(fruits, size=(3, 3))
print(fruit)
```
